there are two types of feature scaling


```python
# standerization
# X′=(X−μ​)/σ
# so that the mean should be 0 and varinace as 1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dataset=pd.read_csv(r"C:\Users\dell\Documents\excells\data1.csv")
dataset.head(3)
```

its is important to do feature scaling before training  a model

```python
dataset.isnull().sum()
```

```python
# presenting in distplot
sns.distplot(dataset["Landsize"])
plt.show()
```

```python
dataset.describe()
```

we can either do manually or use skitlearn library for scaling

```python
# we doing using sklearn
from sklearn.preprocessing import StandardScaler 
ss=StandardScaler()
ss.fit(dataset[["Landsize"]]) #we using fit and transform seperately for ml trining
```

```python
dataset["Landsize_standard"]=pd.DataFrame(ss.transform(dataset[["Landsize"]]),columns=["x"])
dataset.head(3)
```

```python
dataset.describe
```

```python
# comparingn in distplt
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["Landsize_standard"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["Landsize"])
plt.show()
```
