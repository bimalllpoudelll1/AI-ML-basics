```python
# sciketlearn are used to make model for ML but on these one we will use sklearn to fill up the 
# missing numerical datasets in effective way
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# find the actual missing numerical data
dataset.select_dtypes(include="float64").columns
```

```python
from sklearn.impute import SimpleImputer 
```

```python
# using "SimpleImputer" as a approach to fill up in the mean way
si=SimpleImputer(strategy="mean")
ar=si.fit_transform(dataset[['Distance', 'Car', 'BuildingArea', 'YearBuilt', 'Lattitude','Longtitude']])
# we use fit_transform to fill the missing values in the datasets
# output: to remove the array we can use variable like: ar
# array([[ 2.50000000e+00,  1.00000000e+00,  1.51967650e+02,
    #      1.96468422e+03, -3.77996000e+01,  1.44998400e+02],
    #    [ 2.50000000e+00,  0.00000000e+00,  7.90000000e+01,
    #      1.90000000e+03, -3.78079000e+01,  1.44993400e+02],
    #    [ 2.50000000e+00,  0.00000000e+00,  1.50000000e+02,
    #      1.90000000e+03, -3.78093000e+01,  1.44994400e+02],
    #    ...,
    #    [ 6.80000000e+00,  4.00000000e+00,  1.51967650e+02,
    #      1.99700000e+03, -3.78527400e+01,  1.44887380e+02],
    #    [ 6.80000000e+00,  5.00000000e+00,  1.57000000e+02,
    #      1.92000000e+03, -3.78590800e+01,  1.44892990e+02],
    #    [ 6.30000000e+00,  1.00000000e+00,  1.12000000e+02,
    #      1.92000000e+03, -3.78118800e+01,  1.44884490e+02]],
    #   shape=(13580, 6))
```

```python
# to fill the missing values in the columns

pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
```
