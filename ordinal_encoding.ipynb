```python
# another type of encoding is ordinal encoding
# To perform this encoding we have 2 ways. first is with skitlearn 2nd is map
import pandas as pd
df=pd.DataFrame({"size":["m","l","s","xl","s","l","m","s","l","m"]})
df
```

```python
ord_data=[["s","m","l","xl"]]
from sklearn.preprocessing import OrdinalEncoder
ord=OrdinalEncoder(categories=ord_data)
ord.fit(df[["size"]])
 
```

```python
en_size=ord.transform(df[["size"]])
# encoded data
df
```

```python
# for map
ord_mapdata={"s":5,"m":77,"l":999,"xl":0}
df["sixe_map"]=df["size"].map(ord_mapdata)
df
```
