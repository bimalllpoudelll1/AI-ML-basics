```python
# only usefull when the dataset is empty, here gender isnot empty
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python
en_data=dataset[["Gender"]]
```

```python
# one way to perform one hot encoding is using get_dummies
pd.get_dummies(en_data).info()
```

```python
# to encodding into numerical we will use sklearn with preprocessing 
from sklearn.preprocessing import OneHotEncoder
# (WE CAN USE FIT OR FIT TRANSFROM) what fit_transform dooes is anayze the and 
# use skitlearn algorithm and whatever the result it tranform the data into numericsl data when it numeric data
OHE=OneHotEncoder()#drop="first")#we use drop first for removing 1st column of the dataset
ar=OHE.fit_transform(en_data).toarray()
ar
# sparse matrix has its contans filled with 0 and 1 , to convert the data into array we use "toarray"
```

```python
# to convert the array into dataframe we use following
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male","Unisex"])
```
